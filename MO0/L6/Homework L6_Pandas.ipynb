```python
import pandas as pd

# Actividad 1: Lectura de Datos y Análisis Estadístico

# Leer el archivo CSV
# Mostrar las primeras filas del DataFrame

data = pd.read_csv('airbnb.csv')

print("Primeras filas del DataFrame:")
print(data.head())
```

2. Eliminación de Columnas Irrelevantes
Eliminar Columnas:
* Identifica y elimina las columnas que no son relevantes para el análisis (por ejemplo, columnas como id, host id, license).
* Asegúrate de que las columnas relevantes permanezcan en el DataFrame.

```python
# Actividad 2: Eliminación de Columnas Irrelevantes

# Eliminar columnas no relevantes
columnas_a_eliminar = ['id', 'host id', 'license']
clean_data = data.drop(columns=columnas_a_eliminar)

# Verificar columnas restantes
print("Columnas después de eliminación:")
print(clean_data.columns)

```

3. Manejo de Valores Nulos
Eliminar Valores Nulos:
* Utiliza el método dropna() para eliminar filas que contengan valores nulos.
* Verifica que los valores nulos hayan sido eliminados correctamente.

```python
# Actividad 3: Manejo de Valores Nulos

data_null = data[data.isnull().any(axis=1)]
print('filas con datos nulos ')
print(data_null.count())
print(data_null)

no_null_data = data.dropna()

# Eliminar filas con valores nulos

# Verificar si hay valores nulos restantes
print("Valores nulos restantes:")
print(no_null_data)
```

4. Resúmenes con groupby
Agrupar y Resumir Datos:
* Agrupa los datos por neighbourhood group y calcula la cantidad de por cada Barrio 
* Muestra los resultados de esta agrupación.

```python
# Actividad 4: Resúmenes con `groupby`
group_by_neighbourhood_1 = data['neighbourhood group'].value_counts() # internamente es mas rapida porque solo se ddica a oconteo 
# Agrupar por 'neighbourhood group' y resumir precio y número de reseñas
print("Resumen por 'neighbourhood group' con value_counts:")
print(group_by_neighbourhood_1)
```

```python
group_by_neighbourhood_2 = data.groupby('neighbourhood group').size() # grpoup by es mas flexible porque se puede comnbiamr 
# Agrupar por 'neighbourhood group' y resumir precio y número de reseñas
print("Resumen por 'neighbourhood group' con groupby:")
print(group_by_neighbourhood_2)
```

```python
# limpio signo pesos 
clean_data['price'] = clean_data['price'].replace('[\$,]', '', regex=True).astype(float)
group_by_neighbourhood_2 = clean_data.groupby('neighbourhood group').agg(
    mean_price= ('price','mean'),
    mean_review_rate=('review rate number','mean')
).reset_index()

# reset index            precio_medio es para eviatar el cambio de inedex que aplica pandas al agrupar 
###barrio                 
##Palermo           120.0
#Belgrano          100.5


#con reset index  
#barrio  precio_medio
#0   Palermo         120.0
#1  Belgrano        100.5

group_by_neighbourhood_2['mean_price'] = group_by_neighbourhood_2['mean_price'].round(2)
group_by_neighbourhood_2['mean_review_rate'] = group_by_neighbourhood_2['mean_review_rate'].round(0)
# Agrupar por 'neighbourhood group' y resumir precio y número de reseñas
print("Resumen por 'neighbourhood group' con groupby:")
print(group_by_neighbourhood_2)
```

5. Filtrado y Exportación de Datos
Filtrar Datos por País y Año de Construcción:

* Filtra el DataFrame para incluir solo las propiedades en los Estados Unidos (country == 'United States') y construidas después del año 2005.
Guarda este DataFrame filtrado en una nueva variable df_filtrado.
Exportar Datos Filtrados:

* Exporta el DataFrame df_filtrado a un nuevo archivo CSV llamado airbnb_filtrado.csv

```python
# Actividad 5: Filtrado y Exportación de Datos
filter = (data['country'] == 'United States') & (data['Construction year'] > 2015)
filtered_by_country_and_age = data[filter]

# Filtrar por país y año de construcción
print('filtrado por pais y año de construccion')
filtered_by_country_and_age
```

```python
# Exportar el DataFrame filtrado a un nuevo archivo CSV
filtered_by_country_and_age.to_csv('country_age_result.csv', index=False)  
print("Datos filtrados exportados a 'country_age_result.csv'.")
```
