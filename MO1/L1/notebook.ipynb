# Introducción a la Ingeniería de Datos

<p>La Ingeniería de Datos es la disciplina fundamental que permite el desempeño óptimo de otros roles, ya que se encarga del diseño, construcción, mantenimiento y optimización de los sistemas que recopilan, transforman y distribuyen datos de manera eficiente y segura.</p>



# Herramientas 

🔹 ¿Qué es Apache Airflow?
Apache Airflow es una herramienta de orquestación de workflows desarrollada por Airbnb y luego donada a la Apache Foundation. Sirve para:

Automatizar, programar y monitorear flujos de trabajo (workflows).

Definir tareas como código (usando Python).

Ejecutar tareas en un orden específico y con dependencias entre ellas.

Ejemplo de uso: Procesamiento de datos, ETL (Extract, Transform, Load), entrenamiento de modelos de ML, etc.

🧠 ¿Qué es Apache Spark?
Apache Spark es un motor de procesamiento de datos distribuido. Sirve para:

Procesar grandes volúmenes de datos (big data).

Ejecutar transformaciones complejas y consultas (similar a SQL, pero en clústeres).

Trabajar con datos en memoria (muy rápido).

Realizar ETL, machine learning, análisis de grafos, etc.

🗂️ ¿Qué es Apache Hadoop?
Apache Hadoop es un framework de procesamiento de big data distribuido que incluye:

HDFS (Hadoop Distributed File System): un sistema de archivos distribuido.

YARN (Yet Another Resource Negotiator): para gestionar recursos en clústeres.

MapReduce: el motor original de procesamiento por lotes.

📌 Hadoop fue la base de la primera ola de big data, pero hoy en día muchas arquitecturas modernas prefieren usar Spark en lugar de MapReduce porque es mucho más rápido (procesa en memoria).

| Característica     | **Apache Airflow**              | **Apache Spark**            | **Apache Hadoop**                   | **Kubernetes**                        |
| ------------------ | ------------------------------- | --------------------------- | ----------------------------------- | ------------------------------------- |
| ¿Qué hace?         | Orquesta tareas                 | Procesa datos en memoria    | Procesa datos en disco (MapReduce)  | Orquesta contenedores                 |
| ¿Para qué sirve?   | Automatizar flujos y pipelines  | Análisis rápido de big data | Procesamiento por lotes de big data | Desplegar, escalar y administrar apps |
| Lenguaje común     | Python                          | Scala, Python, SQL          | Java, Pig, Hive                     | YAML, Docker                          |
| Motor de ejecución | N/A (depende de otros sistemas) | Interno distribuido         | MapReduce                           | Container runtime (Docker, etc.)      |
| Manejo de datos    | No procesa datos directamente   | En memoria (rápido)         | En disco (lento)                    | No procesa datos directamente         |
| File system        | Usa el que le des (S3, HDFS...) | Puede usar HDFS, S3, etc.   | HDFS                                | No incluye sistema de archivos        |


🔄 ¿Cómo se relacionan?
Airflow puede lanzar jobs de Spark o MapReduce (de Hadoop).

Spark puede correr sobre Hadoop YARN o sobre Kubernetes.

Hadoop define su propio sistema de archivos (HDFS), pero Spark puede usarlo.

Hoy, muchas arquitecturas modernas reemplazan el stack clásico de Hadoop por algo así:

Almacenamiento: S3 (u otro objeto storage)

Procesamiento: Spark

Orquestación: Airflow

Infraestructura: Kubernetes
