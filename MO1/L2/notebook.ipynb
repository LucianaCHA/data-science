# cioclo de vida de datos

Fases generales del ciclo de vida

Es importante recordar que los datos tienen un ciclo de vida, es decir, un conjunto de etapas que atraviesan desde su creación o adquisición hasta su eliminación. Comprender este ciclo es clave para garantizar la calidad, trazabilidad y gobernanza de los datos. Independientemente del tipo de análisis, este ciclo de vida incluye ciertas etapas:

1* 
Generación → Se da cuando en alguno de los sistemas se ingresan datos nuevos.

2 *
Almacenamiento → Se realiza cuando esos datos que se ingresaron se cargan en datalakes o datawarehouses.

3*
Procesamiento → Implica tomar los datos almacenados e introducirlos en nuestros pipelines para realizar las transformaciones necesarias. Estas transformaciones pueden incluir normalizar nombres, validar el tipo de datos cargado, y gestionar valores nulos.

4*
Análisis → Es la etapa en la que los datos procesados se utilizan para diversos fines, como el análisis de ventas, la predicción de tendencias y la implementación en sistemas de inteligencia artificial.

5
Consumo → Este es el punto donde la información generada se pone a disposición para su uso a través de herramientas como dashboards o informes que respaldan la toma de decisiones de los mandos medios y superiores.

6
Eliminación → Se da cuando los datos que tenemos ya no son útiles, por ejemplo, un tablero de las entregas que se hacen en el día. Tenemos la opción de eliminarlos totalmente o hacer una copia de estos y guardarlos en otra ubicación (eliminando los “originales”).

Metadatos: nombre del campo, tipo de dato, descripción, ejemplos, reglas de validación.

Gobernanza de datos

Para establecer la gobernanza de datos en la organización es necesario que se establezcan las políticas, procedimientos, roles y responsabilidades que la regirán. Una vez establecidos, estos puntos guiarán la gestión, seguridad, calidad, integridad y la usabilidad de los datos garantizando un uso efectivo y eficiente de los mismos, siguiendo los objetivos del negocio y regulaciones que los afecten.

# 📚 Buenas Prácticas de Documentación

La documentación, aunque muchas veces subestimada, es clave para solucionar problemas, mantener sistemas y facilitar la incorporación de nuevos miembros al equipo. No siempre implica escribir extensos manuales: puede adoptar múltiples formatos como catálogos de datos, diccionarios, glosarios o diagramas.

---

## 🗂️ Tipos de Documentación Útiles

### 📋 Catálogo de Datos
- Centraliza descripciones, propietarios y **linaje** de los datos.
- El linaje muestra el recorrido del dato: origen, transformaciones, y uso en dashboards, modelos, reportes, etc.

### 📖 Diccionario de Datos
- Incluye, por cada tabla:
  - Nombre del campo
  - Tipo y formato de datos
  - Descripción y posibles valores
  - Valoración (por ejemplo, si es clave, sensible, etc.)
- Puede complementarse con diagramas UML (como el diagrama de clases) para representar relaciones entre tablas.

### 🧾 Glosario de Términos
- Contiene palabras o siglas con su significado.
- Ejemplo:  
  - *Campo* en UI → input en un formulario.  
  - *Campo* en datos → columna de una tabla.

---

## 📌 Buenas Prácticas Generales

- Documentar **durante** el desarrollo del proyecto, no al final.
- Unificar **estilo y formato** para toda la documentación.
- Asegurar **acceso colaborativo** a la documentación desde el inicio.
- Incluir **metadatos clave**:
  - Número de versión
  - Fecha de creación
  - Fecha de última actualización

- Mantenerla **actualizada** mediante revisiones periódicas.
